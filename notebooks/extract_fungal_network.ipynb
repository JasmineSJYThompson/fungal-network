```python
from scipy.io import loadmat 
data=loadmat("Pp_M_Tokyo_U_N_26h_1.mat")
print(data.keys())
```

```python
import pandas as pd
```

```python
A = data["A"]
coordinates = data["coordinates"]
```

```python
A = A.tocoo()
```

```python
print(A.shape)
```

```python
print(coordinates.shape)
```

```python
edges = pd.DataFrame({"source": A.row,
                      "target": A.col,
                      "weight": A.data})
```

```python
edges.head()
print(edges.shape)
```

```python
edges.to_csv("sample-network.csv", index=False)
```
